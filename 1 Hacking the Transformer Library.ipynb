```python
from transformers import pipeline

# Load zero-shot classification pipeline
classifier = pipeline("zero-shot-classification")

# Define candidate labels
candidate_labels = ["politics", "sports", "technology", "entertainment"]

# Define text to classify
text = "The latest advancements in AI and machine learning"

# Perform zero-shot classification
result = classifier(text, candidate_labels)

# Print results
print("Text:", text)
print("Predicted labels:", result['labels'])
print("Scores:", result['scores'])
```

```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load pre-trained model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Text to encode
text = "Once upon a time, there was a kingdom."

# Tokenize input text
inputs = tokenizer.encode(text, return_tensors="pt")

# Generate output
outputs = model.generate(inputs, max_length=50, num_return_sequences=3, do_sample=True)

# Decode and print the generated sequences
for i, output in enumerate(outputs):
    print(f"Generated sequence {i+1}: {tokenizer.decode(output, skip_special_tokens=True)}")
```
